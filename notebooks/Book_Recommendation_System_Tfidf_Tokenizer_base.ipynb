```python
import pandas as pd
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
book_data = pd.read_csv('book30-listing-train.csv',encoding='latin1')
```

```python
s = pd.Series([i for i in range(len(book_data))])
#book_data['index'] = range(1, len(book_data) + 1)
#book_data['index'] = book_data.index

inds = pd.Series(s, index=s)
```

```python
book_data.insert(0, 'index',inds.values)
```

```python
book_data.columns  = ['index','asin', 'filename', 'image_url', 'title', 'author', 'category_id', 'category']
```

```python
 book_data.head()
```

```python
book_data.columns
```

```python
# Selecting the relevant feature for the recommendation
selected_features = ['title','author', 'category']
```

```python
# Replacing the null values with null string 
for feature in selected_features:
    book_data[feature] = book_data[feature].fillna('')
```

```python
book_data['author']
```

```python
# Combining all the 3 selected features 
combined_features = book_data['title']+' '+book_data['author']+' '+book_data['category']  
```

```python
# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# getting a similarity score using consine similarity 
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity.shape)
```

```python
book_name = input('Enter your favourite book name: ')
```

```python
# creating a list with all the book names given in the dataset
list_of_all_books = book_data['title'].tolist()
print(list_of_all_books[:10]) # grap the first 10 book since the Notebook limits
```

```python
# finding the close match for the book name given by the user
find_close_match = difflib.get_close_matches(book_name, list_of_all_books)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# find the index of the book with book name
index_of_the_book = book_data[book_data.title == close_match]['index'].values[0]
print(index_of_the_book)
```

```python
# getting a list of similar books 

similarity_score = list(enumerate(similarity[index_of_the_book]))

print(similarity_score)
```

```python
len(similarity_score)
```

```python
# Sorting the books based on their similarity score 

sorted_similar_books = sorted(similarity_score, key = lambda x:x[1], reverse=True) # x[1] the second element of the list (score value)
print(sorted_similar_books)
```

```python
# print the name of similar movies based on the index 

print('5 books suggest for you : \n ')

i = 1
for book in sorted_similar_books:
    
    
    index = book[0]
    title_from_index = book_data[book_data.index == index]['title'].values[0]
    if (i<6):
        print(i, '.', title_from_index)
        i +=1 
```

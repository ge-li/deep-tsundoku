# Image Cropper

In this notebook, we experiment with the book spine segmentation module in `src/models/image_segmentation.py` and visualize how well it can identify book spines in an image.

```python
import os

import cv2
from PIL import Image

from src.models.image_segmentation import crop_book_spines_in_image, opencv_image_to_pil_image
```

## Identify the book spines of a specific image

```python
image_name = "IMG_1770"
image_ext = "jpg"
image_path = os.path.join("images", f"{image_name}.{image_ext}")

pil_image = Image.open(image_path)
display(pil_image)
```

```python
# Detect book spines in images
book_spines = crop_book_spines_in_image(pil_image, output_img_type="opencv")
print(f"Found {len(book_spines)} book spines 📚")

# Save each book spine as an image
os.makedirs("cropped_images", exist_ok=True)
for i, img in enumerate(book_spines):
    path = os.path.join("cropped_images", f"{image_name}_{i+1}.png")
    display(opencv_image_to_pil_image(img))
    print(f"Saving book spines {i+1} to {path}")
    cv2.imwrite(path, img)
```

## Identify book spines of all images inside the `images` folder

```python
for root, dirs, files in os.walk("images"):
    for filename in files:
        if filename.endswith(".jpg"):
            image_name = filename[:-4]
            image_path = os.path.join("images", f"{image_name}.jpg")
            pil_image = Image.open(image_path)
            book_spines = crop_book_spines_in_image(pil_image, output_img_type="opencv")
            for i, img in enumerate(book_spines):
                cv2.imwrite(
                    os.path.join("cropped_images", f"{image_name}_{i+1}.png"),
                    img
                )
            print(f"{filename} - cropped")
```

```python
# List all book spine images
for root, dirs, files in os.walk("cropped_images"):
    for filename in files:
        print(f"{filename}")
```
